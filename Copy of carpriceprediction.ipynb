```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

```python
data=pd.read_csv('/content/car data.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
#caategorical data
print(data.Fuel_Type.value_counts())
print(data.Seller_Type.value_counts())
```

```python
print(data.Transmission.value_counts())
```

```python
#encoding "fuel type colum
data.replace({'Fuel_Type':{'Petrol':0,'Diesel':1,'CNG':2}},inplace=True)
```

```python
#encoding "transmission column"
data.replace({'Transmission':{'Manual':0,'Automatic':1}},inplace=True)
```

```python
#encoding "seller type"
data.replace({'Seller_Type':{'Dealer':0,'Individual':1}},inplace=True)
```

```python
data.head()
```

```python
X=data.drop(['Car_Name','Selling_Price'],axis=1)
Y=data['Selling_Price']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,random_state=2)
```

```python
#model training
#linear regression
lin_reg=LinearRegression()
```

```python
lin_reg.fit(X_train,Y_train)
```

```python
train_data_pred=lin_reg.predict(X_train)
```

```python
#r square error
error_score=metrics.r2_score(Y_train,train_data_pred)
print("R squared value:",error_score)
```

```python
#visualize the actal price vs predicted price
plt.scatter(Y_train,train_data_pred)
plt.xlabel("ACTUAL PRICE")
plt.ylabel("PREDICTED PRICE")
plt.title("ACTUAL PRICE  V/S  PREDICTED PRICE")
plt.show()
```

```python
test_data_pred=lin_reg.predict(X_test)
error_score=metrics.r2_score(Y_test,test_data_pred)
print("R squared value:",error_score)
```

```python
#visualize the actal price vs predicted price
plt.scatter(Y_test,test_data_pred)
plt.xlabel("ACTUAL PRICE")
plt.ylabel("PREDICTED PRICE")
plt.title("ACTUAL PRICE  V/S  PREDICTED PRICE")
plt.show()
```

```python
#Lasso regression
lass_reg=Lasso()
```

```python
lass_reg.fit(X_train,Y_train)
```

```python
train_data_pred=lass_reg.predict(X_train)
```

```python
#r square error
error_score=metrics.r2_score(Y_train,train_data_pred)
print("R squared value:",error_score)
```

```python
#visualize the actal price vs predicted price
plt.scatter(Y_train,train_data_pred)
plt.xlabel("ACTUAL PRICE")
plt.ylabel("PREDICTED PRICE")
plt.title("ACTUAL PRICE  V/S  PREDICTED PRICE")
plt.show()
```

```python
test_data_pred=lass_reg.predict(X_test)
```

```python
#r square error
error_score=metrics.r2_score(Y_test,test_data_pred)
print("R squared value:",error_score)
```

```python
#visualize the actal price vs predicted price
plt.scatter(Y_test,test_data_pred)
plt.xlabel("ACTUAL PRICE")
plt.ylabel("PREDICTED PRICE")
plt.title("ACTUAL PRICE  V/S  PREDICTED PRICE")
plt.show()
```
